```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, c = 'r')  # With c we can change color
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y graph")
```

```python
plt.scatter(x,y, c = 'r', alpha=0.5)  # With alpha we can control intensity
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y graph")
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c = 'r', marker="X")  # With marker we can control shape of data points
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y graph")
plt.grid()
```

# Sinsudial wave

```python
x = np.linspace(1,10, 100)
```

```python
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y , '--r')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sin wave graph")
plt.grid()
```

## Horizontal bar-plot

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color = '#7E10D5')  # Instead of bar we have to write barh
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

# Histogram

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data,color="#700EBE")
plt.xlabel("representing uniqe val")
plt.ylabel("representing feq")
plt.title("Hist plot example")
plt.grid()
plt.savefig("test.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c='#9327E8',alpha = .9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")

# z label is 
plt.show()
```
